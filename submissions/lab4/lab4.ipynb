```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("multiple_linear_regression_dataset.csv")

print("First 5 rows:\n")
print(data.head())

print("\nColumn names:", data.columns.tolist())
print("Dataset shape:", data.shape)

# Inputs: YearsExperience, TestScore
# Output: Salary
# Model handles 2 features.
```

```python
X = data[["age", "experience"]].values
y = data["income"].values

print("\nShape of X:", X.shape)
print("Shape of y:", y.shape)

# X has 2 columns (2 features)
# y is one numeric target
```

```python
# Normalization prevents large gradient jumps
X_mean = X.mean(axis=0)
X_std = X.std(axis=0)

X = (X - X_mean) / X_std
```

```python
n_features = X.shape[1]

w = np.zeros(n_features)
b = 0.0

# One weight per feature.
# Bias shifts prediction line.
# Starting small keeps learning stable.
```

```python
def predict(X, w, b):
    return X.dot(w) + b

# No activation function because this is regression.
```

```python
def mean_squared_error(y, y_hat):
    return np.mean((y_hat - y) ** 2)

# Squaring makes large income prediction errors expensive.
```

```python
def compute_gradients(X, y, y_hat):
    
    N = len(y)
    
    dw = (2 / N) * X.T.dot(y_hat - y)
    db = (2 / N) * np.sum(y_hat - y)
    
    return dw, db

# Gradient = direction of maximum increase.
# Subtracting it reduces loss.
# If error becomes zero → gradients become zero.
```

```python
learning_rate = 0.01
epochs = 1000


for epoch in range(epochs):
    
    # Forward
    y_hat = predict(X, w, b)
    
    # Loss
    loss = mean_squared_error(y, y_hat)
    
    # Gradients
    dw, db = compute_gradients(X, y, y_hat)
    
    # Update
    w -= learning_rate * dw
    b -= learning_rate * db
    
    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {loss}")

```

```python
print("Final Weights:", w)
print("Final Bias:", b)
```

```python
# prediction for new data point
new_person = np.array([35, 6])

# Normalize using training mean & std
new_person = (new_person - X_mean) / X_std

predicted_income = predict(new_person, w, b)

print("\nPredicted Income:", predicted_income)
```
