```python
import numpy as np
```

```python
X = np.linspace(0, 10, 50).reshape(-1, 1)

# X represents our single input variable.
# Range 0–10 lets us see saturation behavior clearly.
```

```python
# Create nonlinear target with small noise
# Log grows fast first, then slows down

noise = np.random.normal(0, 0.2, size=(50, 1))
y = np.log(X + 1) + noise

# A straight line cannot fit this shape well
# because slope changes across the range.
```

```python
# We choose 3 hidden units.

# Why more than 1?
# One unit cannot bend enough.

# Why not too many?
# Too many units increase instability and overfitting.
```

```python
# Initialize weights and biases

# Input (1) → Hidden (3)
W1 = np.random.uniform(-1, 1, size=(1, 3))
b1 = np.zeros((1, 3))

# Hidden (3) → Output (1)
W2 = np.random.uniform(-1, 1, size=(3, 1))
b2 = np.zeros((1, 1))

# Weights control shape (slopes and bending)
# Biases shift the curve up or down.
```

```python
# ReLU activation

def activation(z):
    return np.maximum(0, z)

# Slope of ReLU

def activation_slope(z):
    return (z > 0).astype(float)

# Slope matters because it tells the model
# how parameters should change to reduce loss.
```

```python
learning_rate = 0.01
epochs = 1000

# Learning rate controls step size.
# 0.01 is stable but still allows learning.

# 1000 epochs is enough for convergence
# without being unnecessarily long.
```

```python
for epoch in range(epochs):
    
    # ----- Forward pass -----
    
    z1 = X @ W1 + b1          # linear transformation
    h = activation(z1)        # non-linear bending
    y_hat = h @ W2 + b2       # output prediction
    
    # ----- Loss -----
    
    error = y_hat - y
    loss = np.mean(error ** 2)   # MSE
    
    # Squaring makes large errors matter more.
    
    
    # ----- Backward pass -----
    
    # dL/dy_hat
    dL_dy = 2 * error / len(X)
    
    # Gradients for output layer
    dL_dW2 = h.T @ dL_dy
    dL_db2 = np.sum(dL_dy, axis=0, keepdims=True)
    
    # These depend on hidden values because
    # output comes directly from hidden layer.
    
    # Flow into hidden layer
    dL_dh = dL_dy @ W2.T
    
    # Apply activation slope
    dL_dz1 = dL_dh * activation_slope(z1)
    
    # Activation controls where error can pass.
    # If slope is zero, no learning happens there.
    
    # Gradients for first layer
    dL_dW1 = X.T @ dL_dz1
    dL_db1 = np.sum(dL_dz1, axis=0, keepdims=True)
    
    
    # ----- Update step -----
    
    # Move opposite to gradient (like perceptron update)
    # Reduce error step by step
    
    W1 -= learning_rate * dL_dW1
    b1 -= learning_rate * dL_db1
    W2 -= learning_rate * dL_dW2
    b2 -= learning_rate * dL_db2
    
    
    if epoch % 100 == 0:
        print("Epoch:", epoch, "Loss:", loss)
```

```python
# Linear models cannot bend.
# Hidden layer + activation allows bending.
# Gradients tell us direction to move.
# Backpropagation is just organized slope computation.
```
