```python
import numpy as np
```

```python
# Simple Perceptron implementation

class Perceptron:
    
    def __init__(self, lr=0.1, epochs=20):
        self.lr = lr
        self.epochs = epochs
        self.w = np.zeros(2)  # two inputs
        self.b = 0.0
    
    def predict(self, x):
        z = np.dot(self.w, x) + self.b
        return 1 if z >= 0 else 0
    
    def train(self, X, y):
        
        for epoch in range(self.epochs):
            for i in range(len(X)):
                
                x_i = X[i]
                y_true = y[i]
                
                y_pred = self.predict(x_i)
                
                error = y_true - y_pred
                
                # Update rule
                self.w += self.lr * error * x_i
                self.b += self.lr * error
        
        print("Final Weights:", self.w)
        print("Final Bias:", self.b)
```

```python
def test_gate(model, X, y, gate_name):
    
    print("\nTesting", gate_name)
    
    for i in range(len(X)):
        prediction = model.predict(X[i])
        print("Input:", X[i], 
              "Expected:", y[i], 
              "Predicted:", prediction)
```

```python
# AND dataset
X_and = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

y_and = np.array([0,0,0,1])

model_and = Perceptron(lr=0.1, epochs=20)
model_and.train(X_and, y_and)

test_gate(model_and, X_and, y_and, "AND")
```

```python
# OR dataset
X_or = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

y_or = np.array([0,1,1,1])

model_or = Perceptron(lr=0.1, epochs=20)
model_or.train(X_or, y_or)

test_gate(model_or, X_or, y_or, "OR")
```
