```python
# basic imports
import numpy as np
import matplotlib.pyplot as plt

# to make results same every time
np.random.seed(42)
```

```python
# generate 100 random values
x = np.random.uniform(-10000, 10000, 100)

# sort values for better plotting
x = np.sort(x)

print("First 10 values:")
print(x[:10])
```

```python
from scipy.special import expit

# Sigmoid (numerically stable)
def sigmoid(x):
    return expit(x)

# Tanh
def tanh(x):
    return np.tanh(x)

# ReLU
def relu(x):
    return np.maximum(0, x)

# Leaky ReLU
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

# ELU
def elu(x, alpha=1):
    return np.where(x > 0, x, alpha * (np.exp(np.clip(x, -100, 50)) - 1))

# Softplus (numerically stable)
def softplus(x):
    return np.where(
        x > 50,
        x,
        np.log(1 + np.exp(np.clip(x, -100, 50)))
    )
```

```python
# Sigmoid derivative
def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)

# Tanh derivative
def tanh_derivative(x):
    return 1 - np.tanh(x)**2

# ReLU derivative
def relu_derivative(x):
    return np.where(x > 0, 1, 0)

# Leaky ReLU derivative
def leaky_relu_derivative(x, alpha=0.01):
    return np.where(x > 0, 1, alpha)

# ELU derivative
def elu_derivative(x, alpha=1):
    return np.where(x > 0, 1, alpha * np.exp(np.clip(x, -50, 50)))

# Softplus derivative
def softplus_derivative(x):
    return sigmoid(x)
```

```python
plt.figure(figsize=(12, 8))

plt.plot(x, sigmoid(x), label="Sigmoid")
plt.plot(x, tanh(x), label="Tanh")
plt.plot(x, relu(x), label="ReLU")
plt.plot(x, leaky_relu(x), label="Leaky ReLU")
plt.plot(x, elu(x), label="ELU")
plt.plot(x, softplus(x), label="Softplus")

plt.title("Activation Functions")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 8))

plt.plot(x, sigmoid_derivative(x), label="Sigmoid Derivative")
plt.plot(x, tanh_derivative(x), label="Tanh Derivative")
plt.plot(x, relu_derivative(x), label="ReLU Derivative")
plt.plot(x, leaky_relu_derivative(x), label="Leaky ReLU Derivative")
plt.plot(x, elu_derivative(x), label="ELU Derivative")
plt.plot(x, softplus_derivative(x), label="Softplus Derivative")

plt.title("Activation Function Derivatives")
plt.legend()
plt.grid(True)
plt.show()
```
